```python
n = 200
x = np.linspace(-1,1,n+1)
y = np.linspace(-1,1,n+1)

a = 0.
b = x[-1]

R = 1
```

```python
def f(x,y,R=1.):
    
    z = R**2 - x**2 - y**2
    
    if z <= 0.:
        return 0.
    else:
        return np.sqrt(z)
    
```

```python
f = np.vectorize(f)
```

```python
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='coolwarm')
```

```python
def integral():
    integ = 0
    for i in range(n):
        for j in range(n):
            promedio = (f(x[j],y[i])+f(x[j+1],y[i])+f(x[j],y[i+1])+f(x[j+1],y[i+1]))/4
            prom_area = (4/40000)*promedio
            integ += prom_area

    return integ
            
        
```
