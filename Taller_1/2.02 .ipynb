Descargue los datos de: https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt Disene un algoritmo
para encontrar todos los maximos locales en esta serie de datos. La Figura [2.1] muestra
la serie de datos y los maximos locales.


```python
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import argrelextrema
url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'
filename = 'EstrellaEspectro.txt'
urllib.request.urlretrieve(url, filename)
data = np.loadtxt(filename)
indicesmax=argrelextrema(data[:, 1], np.greater)[0]
cordenadasmax=data[indicesmax]
plt.plot(data[:, 0], data[:, 1],)
plt.scatter(cordenadasmax[:, 0], cordenadasmax[:, 1], color='red', label='Máximos')
plt.xlabel('[Figura 2.1]: Serie de datos.', labelpad=20)
plt.legend()
plt.show()
```
