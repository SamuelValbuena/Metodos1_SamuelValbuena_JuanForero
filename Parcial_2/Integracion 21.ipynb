Usando las relaciones de ortogonalidad de los polinomios de Legendre,
muestre que los coeficientes est´an dados por

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp

def function(x):
    if x < 0:
        return -1
    elif x > 0:
        return 1
    else:
        return 0
vec_function = np.vectorize(function)
x = np.arange(-1, 1, 0.01)
P = np.polynomial.legendre.legval(x, np.eye(21))
N = 20
Coefs = np.zeros(N + 1)
for n in range(N + 1):
    Coefs[n] = (2 * n + 1) / 2 * np.trapz(vec_function(x) * P[n], x)

f_Leg = np.polynomial.legendre.legval(x, Coefs)

n = 15
x_GL, w_GL = np.polynomial.legendre.leggauss(n + 1)

P_GL = np.polynomial.legendre.legval(x_GL, np.eye(N + 1))

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(x, vec_function(x), label='Función original')
ax[0].plot(x, f_Leg, label='Aproximación en polinomios ')
ax[0].legend()
ax[0].set_xlabel('x')
ax[0].set_ylabel('f(x)')
ax[0].set_title('Aproximación en polinomios')
plt.axis('off')
plt.show()
```
